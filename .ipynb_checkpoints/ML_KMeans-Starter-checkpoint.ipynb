## Importando librerías

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets.samples_generator import make_blobs
%matplotlib inline
```

```python
np.random.seed(0)
```

```python
# Definiendo las muestras
X, y = make_blobs(n_samples=5000,
centers=[[4,4], [-2,-1], [2,-3], [1,1]],

cluster_std = 0.9)
```

```python
plt.scatter(X[:,0], X[:,1], marker="*")
```

```python
k_means = KMeans(init = "k-means++", n_clusters=4, n_init=12)
```

```python
#help(KMeans)
```

```python
k_means.fit(X)
```

```python
k_means_labels = k_means.labels_
k_means_labels
```

```python
k_means_cluster_centers = k_means.cluster_centers_
k_means_cluster_centers
```

```python
# Iniciamos el gráfico con estas dimensiones 6 y 4
fig = plt.figure(figsize=(6, 4))

# Colores usa un mapa de colores, lo que nos producirá un arreglo de colores
# basado en el número de etiquetas que tenemos. Usamos set(k_means_labels)
# para obtener etiquetas únicas
colors = plt.cm.Spectral(np.linspace(0, 1, len(set(k_means_labels))))

# Creando el gráfico
ax = fig.add_subplot(1, 1, 1)

# Bucle que grafica los puntos de datos y sus centroides
# k estará en un rango de 0-3, lo que coincidirá los posibles grupos
# que tienen los puntos
for k, col in zip(
    range(len([[4,4], [-2, -1], [2, -3], [1, 1]])), 
    colors):

    # Crea una lista de todos los datos, donde los puntos de datos
    # que están en el grupo. (ex: grupo 0) son etiquedas como verdadero
    # y sino son como falso
    my_members = (k_means_labels == k)

    # Define el centroide, o el centro del grupo
    cluster_center = k_means_cluster_centers[k]

    # Grafica los datos con el color col.
    ax.plot(X[my_members, 0], X[my_members, 1], 'w', 
            markerfacecolor=col, 
            marker='.')
    
    # Grafica los centroides con el color especificado
    # pero con un contorno más oscuro
    ax.plot(cluster_center[0], cluster_center[1], 
            '*', markerfacecolor=col,  
            markeredgecolor='k', 
            markersize=6)

# Título del gráfico
ax.set_title('KMeans')

# Removes  el eje X
ax.set_xticks(())

# Removemos el eje Y
ax.set_yticks(())

# Mostramos el gráfico
plt.show()
```

## Problema

Imagine que tiene un conjunto de datos de clientes y necesita aplicar la segmentación de clientes en estos datos históricos. La segmentación de clientes es la práctica de dividir una base de clientes en grupos de individuos que tienen características similares. Es una estrategia importante ya que una empresa puede apuntar a estos grupos específicos de clientes y asignar efectivamente recursos de marketing. Por ejemplo, un grupo puede contener clientes que son de alto beneficio y bajo riesgo, es decir, más propensos a comprar productos o suscribirse a un servicio. Una tarea comercial es retener a esos clientes. Otro grupo podría incluir clientes de organizaciones sin fines de lucro. Y así.

```python
import pandas as pd
cust_df = pd.read_csv("Cust_Segmentation.csv")
cust_df.head()
```

```python
cust_df.dtypes
```

```python
df = cust_df.drop("Address", axis=1)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

X = df.values[:,1:]
X = np.nan_to_num(X)
Clus_dataSet = StandardScaler().fit_transform(X)
Clus_dataSet
```

## Modelado

```python
clusterNum = 3
k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 12)
k_means.fit(X)
labels = k_means.labels_
print(labels)
```

```python
df.shape
```

```python
df["Clus_km"] = labels
df.head()
```

```python
df.groupby("Clus_km").mean()
```

```python
area = np.pi * (X[:,1]**2)

plt.scatter(X[:,0], X[:,3], s=area, c=labels.astype(np.float), alpha=0.5)
plt.xlabel("Age", fontsize=18)
plt.ylabel("Income", fontsize=18)

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(1, figsize=(8,6))

plt.clf()

ax = Axes3D(fig, rect=[0,0,.95,1], elev=48, azim=134)

plt.cla()
ax.set_xlabel("Education")
ax.set_ylabel("Age")
ax.set_zlabel("Income")

ax.scatter(X[:,1], X[:,0], X[:,3], c= labels.astype(np.float))

```
